```python
import pandas as pd, pathlib, numpy as np, os
path = pathlib.Path("data/processed/cicids2017_ml")
files = sorted(path.glob("*.parquet"))
print("Files found:", len(files))
files
```

```python
dfs = []
for f in files:
    print("Loading:", f.name)
    df = pd.read_parquet(f)
    dfs.append(df)
data = pd.concat(dfs, ignore_index=True)
print("Merged shape:", data.shape)
```

```python
data.columns = [c.strip() for c in data.columns]
print("Sanitised columns:", data.columns[:10])
```

```python
data['Label'] = data['Label'].replace({'BENIGN': 0}).apply(lambda x: 0 if x == 0 else 1)
print("Label distribution:\n", data['Label'].value_counts())
```

```python
before = len(data)
data = data.dropna()
after = len(data)
print(f"Dropped {before - after} rows with NaN values")
```

```python
os.makedirs("data/processed", exist_ok=True)
data.to_parquet("data/processed/cicids2017_clean.parquet", index=False)
print("Saved cleaned dataset -> data/processed/cicids2017_clean.parquet")
```
